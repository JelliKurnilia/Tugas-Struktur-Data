```python
# JAWABAN NO 5 
class Stack:
  def __init__(self):                 # Pendefinisian yang digunakan untuk melakukan inisialisasi pembuatan object dari class tertentu
    self._data = []
    
  def __len__(self):                  # Pendefinisian yang dilakukan agar objek dari kelas stack dapat dihitung jumlah elemennya menggunakan metode len().
    return len(self._data)

  def isempty(self):                  # Digunakan untuk memeriksa apakah stack kosong atau tidak
    return len(self._data) == 0
  
  @property #sebagai decorator
  def view(self):
    return self._data

  def push(self,item):                # Inisialisasi fungsi Push (berguna untuk menambah elemen baru ke dalam stack pada posisi terakhir)
    self._data.append(item)

  def pop(self):                      # Inisialisasi fungsi Pop (berguna untuk mengambil elemen terakhir dan menghapusnya dari dalam stack)
    if self.isempty():
      raise Exception('Stack kosong') # exception berguna agar program tidak dilanjutkan ketika eror
    return self._data.pop()

  def top(self):                      #Inisialisasi fungsi Top (berguna untuk mengecek posisi atas)
    if self.isempty():
      raise Exception('Stack kosong')
    return self._data[-1]
```

```python
def permutasi(arr):       # mendefinisikan fungsi permutasi dengan array
  if len(arr) == 0:       # berfungsi untuk mengembalikan panjang (jumlah anggota array)
    return []             # berfungsi untuk melakukan perulangan (mengembalikan)
  if len(arr) == 1:
    return [arr]
  p = []
  for i in range(len(arr)):
    a = arr[i]
    b = arr[:i]+arr[i+1:]
    for k in permutasi(b):
      p.append([a]+k)      # Append berfungsi untuk menambahkan sesuatu
  return p                 # mereturn/mengulang permutasi
```

```python
data = Stack()                   # Mendefinisikan variabel Data sebagai sebuah stack dengan memasukkan nilai tertentu
for i in range(5, 0, -1):        # Membuat tumpukan dengan range kombinasi nilai tertentu
  data.push(i)                   # Memasukkan elemen ke dalam stack

for i in permutasi(data.view):   # Nilai pada variabel i dicetak dalam bentuk permutasi
  print(i)                       # Mencetak nilai dari variabel i
```

* Facade -> Pola ini dapat mendefinisikan interface yang membuat subsistem lebih mudah dimengerti dan digunakan.

* Iterator -> digunakan untuk mengakses elemen dari collection object secara berurutan tanpa mengetahui representasi yang mendasarinya (list, stack, tree, dll)

* Prototype -> Pola ini memungkinkan pengguna untuk menyalin objek menggunakan instance prototype yang ada dengan membuat objek yang baru dan dapat dikustomisasi sesuai dengan kebutuhan tanpa bergantung dengan objek aslinya.

* Strategy ->  membantu dalam mendefinisikan family algorithm, kemudian mengenkapsulasi masing-masing algoritma ke dalam class yang terpisah dan membuat objek dapat dipertukarkan (objects interchangeable). 
* Template method -> Pola ini memungkinkan kamu untuk mendefinisikan kerangka algoritma ke dalam base class dan membiarkan subclass mendefinisikan kembali langkah-langkah tertentu dari suatu algoritma tanpa mengubah struktur algoritma.
* Visitor -> Pola ini memungkinkan kamu untuk menentukan operasi baru tanpa mengubah kelas elemen tempat operasi itu beroperasi.

```python
# 3. Buatlah ADT untuk Queue dan Deque menggunakan adapter design patter
class Queue:
  def __init__(self):
    self._data = []
    self._size = 0
  
  def __len__(self):
    return self._size

  def isempty(self):
    return len(self._data) == 0
  
  @property
  def view(self):
    return self._data
 
  def first(self):
    return self._data[0]
  
  def last(self):
    return self._data[-1]
  

  def enqueue(self,item):
    self._data.append(item)
    self._size += 1

  def dequeue(self):
    if self.isempty():
      raise Exception('Stack kosong')
    else:
      self._size -= 1
      return self._data.pop(0)
```

```python
class StackWithQueue:
  def __init__(self):
    self._data = Queue()
    self._size = 0

  def __len__(self):
    return self._size

  def isempty(self):
    return self._size == 0
 
  @property
  def view(self):
    return self._data.view
  
  def push(self,item):
    self._data.enqueue(item)
    self._size += 1

    for i in range(self._size -1):
      self._data.enqueue(self._data.dequeue())
  
  def pop(self):
    if self.isempty():
      raise Exception('Stack kosong')
    self._size -= 1
    return self._data.dequeue()

  def top(self):
    return self._data[-1]
```

```python
class Deque:
  def __init__(self):
    self._data = []
    self._size = 0
  
  def __len__(self):
    return self._size

  def isempty(self):
    return len(self._data) == 0
  
  @property
  def view(self):
    return self._data
 
  def first(self):
    return self._data[0]
  
  def last(self):
    return self._data[-1]
  
  def add_first(self,item):
    self._data.append(item)
    self._data.insert(0,self._data.pop())
    self._size += 1
  
  def add_last(self,item):
    self._data.append(item)
    self._size += 1

  def delete_first(self):
    if self.isempty():
      raise Exception('Stack kosong')
    else:
      self._size -= 1
      return self._data.pop(0)

  def delete_last(self):
    if self.isempty():
      raise Exception('queue kosong')
    else:
      self._size -= 1
      return self._data.pop()
```

```python
# 4. Buatlah program untuk mengimplementasikan ADT tersebut menggunakan adapter design pattern
# Dog - Cycle
# human - Truck
# car - Car
 
class MotorCycle:
 
    """Class for MotorCycle"""
 
    def __init__(self):
        self.name = "MotorCycle"
 
    def TwoWheeler(self):
        return "TwoWheeler"
 
 
class Truck:
 
    """Class for Truck"""
 
    def __init__(self):
        self.name = "Truck"
 
    def EightWheeler(self):
        return "EightWheeler"
 
 
class Car:
 
    """Class for Car"""
 
    def __init__(self):
        self.name = "Car"
 
    def FourWheeler(self):
        return "FourWheeler"
 
class Adapter:
    """
    Adapts an object by replacing methods.
    Usage:
    motorCycle = MotorCycle()
    motorCycle = Adapter(motorCycle, wheels = motorCycle.TwoWheeler)
    """
 
    def __init__(self, obj, **adapted_methods):
        """We set the adapted methods in the object's dict"""
        self.obj = obj
        self.__dict__.update(adapted_methods)
 
    def __getattr__(self, attr):
        """All non-adapted calls are passed to the object"""
        return getattr(self.obj, attr)
 
    def original_dict(self):
        """Print original object dict"""
        return self.obj.__dict__
 
 
""" main method """
if __name__ == "__main__":
 
    """list to store objects"""
    objects = []
 
    motorCycle = MotorCycle()
    objects.append(Adapter(motorCycle, wheels = motorCycle.TwoWheeler))
 
    truck = Truck()
    objects.append(Adapter(truck, wheels = truck.EightWheeler))
 
    car = Car()
    objects.append(Adapter(car, wheels = car.FourWheeler))
 
    for obj in objects:
       print("A {0} is a {1} vehicle".format(obj.name, obj.wheels()))
```
